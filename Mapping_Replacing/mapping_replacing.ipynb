Mapping and Replacing

Mapping refers to converting values from one form into another form. Then replacing means changing the old values into the new values.

```python
# Create a copy of the dataframe
map_df = df.copy()
# Check the columns for the dataframe
map_df.columns
```

1 Mapping values with .map()

Use the function of .map() when each old value has one matching new value. The argument that pass through the function is a dictionary.

```python
# Create the dictionary
department_map = {
    "HR": "Human Resources",
    "IT": "Information Technology",
    "FIN": "Finance",
    "MKT": "Marketing",
    "OPS": "Operations"
}
# Map the dicitonary
map_df["department_full"] = map_df["department_code"].map(department_map)

# Check the dataframe
map_df.head()
```

2 Replacing values with .replace()

Using the .replace() function when you want to change the values inside a column. This is done by also passing a dictionary into the function.

```python
# Create the dicitonary
replace = {
    "Avg" : "Average",
    "Needs Improvement": "Poor"
}

# Use the .replace() to replace the values
map_df['performance_rating'] = map_df['performance_rating'].replace(replace)

map_df.head()
```

3 Standardizing text before mapping

There are cases that the values would not match because there is either spacing or capitals, in order to fix you need to standardize it. You do this by using the .str.strip() and then .str.lower()

```python
# Stardardize the department name column
map_df['department_name'] = map_df['department_name'].str.strip().str.lower()

# Check the dataframe 
map_df.head()
```

4 Mapping number from categories

Mapping number from categories is possible by using the dictionary and having the second value be a number.

```python
# Create a dictionary that is able to change 
# the performance rating into numeric values
rating_score = {
    "Excellent": 4,
    "Good": 3,
    "Average": 2,
    "Poor": 1
}

# Map to the performance_rating
map_df["rating_score"] = map_df["performance_rating"].map(rating_score)

# Check the dataframe
map_df.head()
```

5 Using .fillna() after .map()

Sometimes when you using the .map() there can create a bunch NAs in the dataframe to avoid that you can use the .fillna() function.

```python
# Use the previous dictionary, to map the departments
map_df["department_full"] = (
    map_df["department_code"]
    .map(department_map)
    .fillna("Unknown")
)

# Check the dataframe
map_df.head()
```

6 Replace partial text with .str.replace()

In order to replace part of a string, you can use the function of .str.replace(), and the pass arguments of "old_text", "new_text", and then regex=False.

```python
# For the departname column replace the Dept with Department
map_df['department_name'] = map_df['department_name'].str.replace(
    "Dept",
    "Department",
    regex = False
)
# Check the dataframe
map_df.head()
```

7 Replace values across mulitple columns

Instead of applying the replace function to only one column to can pass in more columns to apply it to many, so it uses the same .replace() but when you are filtering use a list to pass more columns.

```python
# For the employment type and remote status replace key words
map_df[["employment_type", "remote_status"]] = (
    map_df[["employment_type", "remote_status"]]
    .replace({
        "Temp": "Temporary",
        "contractor": "Contract",
        "yes": "Yes",
        "no": "No"
    })
)
# Check the datafame
map_df.head()
```

8 Using .replace() across the whole DataFrame

Instead of replacing by the specific columns you can just use the function on the whole entire dataframe instead. 

```python
# For the whole dataframe replace all the Y and N to yes and no
map_df = map_df.replace(
    {
        "Y" : "Yes",
        "N" : "No"
    }
)

# Check the dataframe
map_df.head()
```

10 Using .rename() to replace column names

In orde to repalce the columns you can use the function of .rename(), which this takes dictionary of the pre_name and then the next name

```python
# Replace the column names of emnployee_id with emp_id, 
# and then employee_name with name
map_df = map_df.rename(columns={
    "employee_id": "emp_id",
    "employee_name": "name"
})
# Check the dataframe
map_df.head()
```

11 Using .where() to replace values conditionally

In order to use the .where() function you pass by a condition and the replacement value, this basically keeps the value where the condition is true, and then replaces the values where the condition is false.

```python
# For the salary band column check to see if the values are 
# A, B, C, D if not its unknown
map_df["salary_band_checked"] = map_df["salary_band"].where(
    map_df["salary_band"].isin(["A", "B", "C", "D"]),
    "Unknown"
)

# Check the dataframe
map_df.head()
```

12 Using .mask() to replace values conditionally 

The .mask() function is the complete opposite of the .where() function, as it uses the condition and replacement value, but when the condition is true you replace and if the condition is false you keep.

```python
# In the performance rating column find the 
# values that have poor, and replace with review

map_df["performance_rating_checked"] = map_df["performance_rating"].mask(
    map_df["performance_rating"] == "Poor",
    "Needs Review"
)

# Check the dataframe
map_df.head()
```
