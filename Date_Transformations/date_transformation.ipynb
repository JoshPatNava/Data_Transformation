Date Transformations


```python
# Copy the data
date_df = df.copy()
# Check the data
date_df.columns
```

1 Convert text dates using the pd.to_datetime()

At the start of the date transformations you have to change the date column using the pd.to_datetime()

```python
# Convert that date column into a datetime

# Create a list of the columns
date_columns = ["event_date", "start_date", "end_date", "last_active_date", "reminder_date"]
# Convert all of those columns into datetime
for col in date_columns:
    date_df[col] = pd.to_datetime(date_df[col])
# Check the data types
date_df.dtypes
```

2 Extract the year

You can extract the year from a column by using .dt.year

```python
# Extract the year from the ever_date column
date_df['Event_Year'] = date_df['event_date'].dt.year
# Check the data frame
date_df[['record_id', 'event_name', 'Event_Year']]
```

3 Extract the month number

To extract the month number you can use the function of .dt.month

```python
# Extract the month out of the event_month column
date_df["event_month"] = date_df["event_date"].dt.month
# Check the data frame
date_df.head()
```

4 Extract the month name 

In order to extract the month name you need to use the funciton of .dt.month_name()

```python
# Extract the month name out of the event_date column
date_df["event_month_name"] = date_df["event_date"].dt.month_name()
# Check the dataframe
date_df.head()
```

5 Extract the day of the week

To extract the day of the week you need to use the .dt.day

```python
# Extract the day from the event_date column
date_df['event_weekday'] = date_df['event_date'].dt.day
# Check the dataframe
date_df.head()
```

6 Extract the quarter

The quarter refers to the range of month, Jan-Mar being Q1 and so on to find this you can use the .dt.quarter

```python
# Extract the quarter from the event_date column
date_df['event_quarter'] = date_df['event_date'].dt.quarter
# Check the dataframe
date_df.head()
```

7 Create Year-Month Periods

To create the year month periods you have to use the function of pd.to_period(), which this could be useful when you are trying to do the monthly trend analysis.

```python
# Find the monthly period for the event_date
date_df["event_year_month"] = date_df["event_date"].dt.to_period("M")
# Check the datafrmae
date_df.head()
```

8 Calculate the difference between dates

To calculate the difference between the dates you can use the minus from the end of the date till the start of it.

```python
# Find the difference betwene the end_date and the start_date
date_df["duration_days"] = (date_df["end_date"] - date_df["start_date"]).dt.days
# Check the dataframe
date_df.head()
```

9 Calculate days since last activity

To calculate the days since last activity you can use the difference between the date of "today" and the when last active.

```python
# Create the "today"
today = pd.to_datetime("2026-06-01")
# Find the difference
date_df["days_since_last_active"] = (
    today - date_df["last_active_date"]
).dt.days
# Check the dataframe
date_df.head()
```

10 Calculate days until reminder

This is done by also using the today and then subtracting it from a reminder date

```python
# Create the "today"
today = pd.to_datetime("2026-06-01")
# Find the days until the reminder
date_df["days_until_reminder"] = (
    date_df["reminder_date"] - today
).dt.days
# Check the dataframe
date_df.head()
```

11 Create a past/future label

This is done by also using a today, and then using a condition to decide whether that date has already past or will be in the future

```python
# Create the today
today = pd.to_datetime("2026-06-01")
# Find whether that date will past or future
date_df["reminder_status"] = np.where(
    date_df["reminder_date"] < today,
    "Past Reminder",
    "Upcoming Reminder"
)
# Check the dataframe
date_df.head()
```

12 Create a duration category

To create this duration category, start with conditions that will decide whether that duration will be short, medium, or long, and then using that create the column using np.select()

```python
# Create the conditions
conditions = [
    date_df["duration_days"] <= 5,
    date_df["duration_days"].between(6, 9),
    date_df["duration_days"] >= 10
]
# Create the categories
choices = ["Short", "Medium", "Long"]
# Create the new column
date_df["duration_category"] = np.select(
    conditions,
    choices,
    default="Unknown"
)
# Check the data frame
date_df.head()
```

13 Sort by Date

You can sort by the date by using the function .sort_values()

```python
# Sort the dataframe by the event_date
date_df.sort_values('event_date')
```

14 Filter by date

You can use filtering methods on date columns if they have the type of date time. This could be done by boolean masking or just .filter()

```python
# Find the event_dates that are more than 2026-03-01
date_df[date_df["event_date"] >= "2026-03-01"]
```

15 Format dates as strings

You can format the dates as strings, by using the .dt.strftime() and then passing the arugment of a specific formatting.

```python
# Format the event date column
date_df["formatted_event_date"] = date_df["event_date"].dt.strftime("%m/%d/%Y")
# Check the data frame
date_df.head()
```
